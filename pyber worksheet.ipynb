```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City  Data
cities_data=pd.read_csv("data/city_data.csv")
cities_data.head()
```

```python
# Read the Ride Data
rides_data=pd.read_csv("data/ride_data.csv")
rides_data.head()
```

```python
# Combine the data into a single dataset and display
cityride_merged=pd.merge(rides_data,cities_data, on='city')
cityride_merged.head()
```

# Bubble plot for ride sharing data

```python
#average fare per city and type
```

```python
#grouping by city and type
city_group=cityride_merged.groupby(["city","type"])
city_group.head()
```

```python
#calculating average fair per city
average_fare=city_group["fare"].mean().reset_index()
```

```python
#setting the index by 'type'
average_fare_by_type=average_fare.set_index('type')
```

```python
# using loc for cal avg fare for urban,rural & suburban
average_fare_urban=average_fare_by_type.loc["Urban"]
average_fare_rural=average_fare_by_type.loc["Rural"]
average_fare_suburban=average_fare_by_type.loc["Suburban"]
```

```python
# cal total number of rides per city
rides_count=city_group['ride_id'].count().reset_index()
rides_count
```

```python
#counting by type
rides_count_by_type=rides_count.set_index('type')
rides_count_by_type
```

```python
#using loc to separate urban,rural and suburban
rides_count_urban=rides_count_by_type.loc["Urban"]
rides_count_urban
```

```python
rides_count_rural=rides_count_by_type.loc["Rural"]
rides_count_rural
```

```python
rides_count_suburban=rides_count_by_type.loc["Suburban"]
rides_count_suburban
```

```python
#total drivers per city
drivers=cities_data[["city","driver_count","type"]]
drivers
```

```python
#counting drivers by type
drivers_by_type=drivers.set_index('type')
drivers_by_type
```

```python
drivers_urban=drivers_by_type.loc["Urban"]
drivers_urban
```

```python
drivers_rural=drivers_by_type.loc["Rural"]
drivers_rural
```

```python
drivers_suburban=drivers_by_type.loc["Suburban"]
drivers_suburban
```

#Scatter plotting data

```python
#plot data for Urban cities

x_axis_urban= rides_count_urban['ride_id']
y_axis_urban= average_fare_urban['fare']
plt.scatter(x_axis_urban,y_axis_urban, color="lightcoral",s=drivers_urban['driver_count']*9,alpha=1.0,linewidths=1,edgecolors='k',
           label="Urban")

#plot data for Suburban cities
x_axis_suburban= rides_count_suburban['ride_id']
y_axis_suburban= average_fare_suburban['fare']
plt.scatter(x_axis_suburban,y_axis_suburban,color="lightskyblue",s=drivers_suburban['driver_count']*6,
            alpha=1.0,linewidths=1,edgecolors='k', label="Suburban")


#plot data for Rural cities
x_axis_rural= rides_count_rural['ride_id']
y_axis_rural= average_fare_rural['fare']
plt.scatter(x_axis_rural,y_axis_rural,color="gold", s=drivers_rural['driver_count']*5,alpha=1.0,linewidths=1,edgecolors='k',
            label="Rural")


plt.legend(loc="best")
plt.xlabel("Total number of Rides(Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data")
plt.show()
```

# Total fares by city type

```python
#total fares from data
total_fares=cityride_merged["fare"].sum()
total_fares
```

```python
#percentage of total fares by city type
city_type=cityride_merged.groupby('type')
total_fares_by_type=city_type.sum().reset_index()
total_fares_by_type
```

```python
#total_fares_by_type

type_index=total_fares_by_type.set_index('type')
type_index
```

```python
rural_fare=type_index.loc[["Rural"],["fare"]]
percentage_rural_fare=((rural_fare["fare"])/(total_fares)*100)
percentage_rural_fare
```

```python
urban_fare=type_index.loc[["Urban"],["fare"]]
percentage_urban_fare=((urban_fare["fare"])/(total_fares)*100)
percentage_urban_fare
```

```python
suburban_fare=type_index.loc[["Suburban"],["fare"]]
percentage_suburban_fare=((suburban_fare["fare"])/(total_fares)*100)
percentage_suburban_fare
```

# Total Drivers by City Type

```python
#total drivers from data
total_drivers=cities_data["driver_count"].sum()
total_drivers
```

```python
#percentage of total drivers by city type
city_type=cities_data.groupby('type')
total_drivers_by_type=city_type.sum().reset_index()

type_index=total_drivers_by_type.set_index('type')
type_index
```

```python
rural_drivers=type_index.loc[["Rural"],["driver_count"]]
percentage_rural_drivers=((rural_drivers["driver_count"])/(total_drivers)*100)
percentage_rural_drivers
```

```python
urban_drivers=type_index.loc[["Urban"],["driver_count"]]
percentage_urban_drivers=((urban_drivers["driver_count"])/(total_drivers)*100)
percentage_urban_drivers
```

```python
suburban_drivers=type_index.loc[["Suburban"],["driver_count"]]
percentage_suburban_drivers=((suburban_drivers["driver_count"])/(total_drivers)*100)
percentage_suburban_drivers
```

# Total Fares by city type

```python
#pie charts
labels=["Rural","Urban","Suburban"]
sizes = [percentage_rural_fare,percentage_urban_fare,percentage_suburban_fare]

colors=["gold","lightskyblue","lightcoral"]
explode=(0,0.1,0)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.axis("equal")
plt.title("Percentage of Total Fares by City Type")
```
